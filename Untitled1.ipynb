```python

import pandas as pd
import numpy as np
import matplotlib.pyplot
```

```python


df = pd.read_csv("mca_gujarat_21042018.csv", encoding='latin1')
```

```python
# Display first few rows
print(df.head())

# Show dataset information
df.info()

# Describe statistics
print(df.describe(include='all'))

```

```python

# Count missing values
print(df.isnull().sum())

df['AUTHORIZED_CAP'].fillna(df['AUTHORIZED_CAP'].mean(), inplace=True)

# Verify
print(df.isnull().sum())
```

```python

stats = df.groupby(['REGISTRAR_OF_COMPANIES', 'COMPANY_CLASS']).size().reset_index(name='Count')
print(stats)

```

```python
import matplotlib.pyplot as plt

stats.pivot(index='REGISTRAR_OF_COMPANIES', columns='COMPANY_CLASS', values='Count').plot(kind='bar')
plt.title("Private vs Public Companies by Registrar (Gujarat)")
plt.xlabel("Registrar of Companies")
plt.ylabel("Number of Companies")
plt.show()
```

```python

# Group by company status
status_group = df.groupby('COMPANY_STATUS').size().reset_index(name='Count')

# Display statistics
print(status_group)

# Visualize
plt.figure(figsize=(8,5))
plt.bar(status_group['COMPANY_STATUS'], status_group['Count'], color='skyblue')
plt.title("Number of Companies by Status")
plt.xlabel("Company Status")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.show()


```

```python
# Group by company status and subcategory
status_sub_group = df.groupby(['COMPANY_STATUS', 'COMPANY_SUB_CATEGORY']).size().reset_index(name='Count')

# Display results
print(status_sub_group.head(10))  # show top 10 groups

# Optional: Visualization
pivot_data = status_sub_group.pivot(index='COMPANY_SUB_CATEGORY', columns='COMPANY_STATUS', values='Count')
pivot_data.plot(kind='bar', figsize=(10,6))
plt.title("Company Count by Status and Subcategory")
plt.xlabel("Company Subcategory")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.show()
```

```python
status_counts = df['COMPANY_STATUS'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', colors=['skyblue','lightgreen','orange'])
plt.title("Percentage of Applicants by Company Status")
plt.show()

```

```python
class_counts = df['COMPANY_CLASS'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%')
plt.title("Percentage of Applicants by Company Class")
plt.show()
```

```python
activity_class = df.groupby(['PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN','COMPANY_CLASS']).size().unstack()
activity_class.plot(kind='bar', figsize=(8,6))
plt.title("Private/Public Companies by Business Activity")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['REGISTRAR_OF_COMPANIES'].value_counts().plot(kind='bar')
plt.title("Companies by Registrar of Companies")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(df['AUTHORIZED_CAP'])
plt.title("Boxplot of Authorized Capital")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['AUTHORIZED_CAP'], df['REGISTRAR_OF_COMPANIES'].astype('category').cat.codes, alpha=0.6)
plt.xlabel("Authorized Capital")
plt.ylabel("Registrar of Companies (Encoded)")
plt.title("Scatter Plot: Authorized Capital vs Registrar")
plt.show()
```

```python
# Example dataset (replace with your actual data)
districts = ['Ahmedabad', 'Surat', 'Rajkot', 'Vadodara', 'Bhavnagar']
population = [7200000, 6000000, 3000000, 4000000, 2500000]
lat = [23.02, 21.17, 22.30, 22.31, 21.77]
lon = [72.57, 72.83, 70.78, 73.18, 72.15]

plt.figure(figsize=(8,8))
m = Basemap(projection='merc', llcrnrlat=20, urcrnrlat=25, llcrnrlon=68, urcrnrlon=75, resolution='i')
m.drawmapboundary(fill_color='lightblue')
m.fillcontinents(color='lightgreen', lake_color='lightblue')
m.drawcountries()
m.drawstates()

x, y = m(lon, lat)
m.scatter(x, y, s=[p/10000 for p in population], color='red', alpha=0.7)

for i, district in enumerate(districts):
    plt.text(x[i]+10000, y[i]+10000, district, fontsize=9)

plt.title("Gujarat Districts Population Distribution")
plt.show()
```
